```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Đọc metadata thu thập từ website https://climatedata.imf.org/pages/climatechange-data 

```python
df_land_cover = pd.read_csv('C:\\Users\\ngmai\\Downloads\\Land_Cover_Accounts (2).csv')
df_land_cover
```

Lọc lại dữ liệu cần cho phân tích (từ năm 2018 đến năm 2022)

```python
df_land_cover = df_land_cover[['Country', 'ISO3', 'Indicator', 'Unit', 'F2018', 'F2019', 'F2020', 'F2021', 'F2022']]
df_land_cover
```

Kết hợp với bộ dữ liệu asian_countries để lọc lại dữ liệu 51 quốc gia Châu Á

```python
asian = pd.read_csv('C:\\Users\\ngmai\\Downloads\\asian_countries (1).csv')
```

```python
asian
```

```python
asian_land_cover = pd.merge(df_land_cover, asian, on='ISO3')
asian_land_cover
```

```python
asian_land_cover = asian_land_cover.drop(['Unnamed: 0', 'Country_y'], axis=1)
```

```python
asian_land_cover = asian_land_cover.rename(columns={"Country_x":"Country", "F2018":"2018", "F2019":"2019", "F2020":"2020", "F2021":"2021", "F2022":"2022"})
asian_land_cover
```
