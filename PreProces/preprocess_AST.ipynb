```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Đọc metadata thu thập từ website https://climatedata.imf.org/pages/climatechange-data 

```python
ast = pd.read_csv('../data/raw_data/annual_surface_temperature_change.csv')
ast
```

Lọc lại dữ liệu cần cho phân tích (từ năm 2018 đến năm 2022)

```python
ast = ast[['Country', 'ISO3', 'Indicator', 'Unit', '2018', '2019', '2020', '2021', '2022']]
ast
```

Kết hợp với bộ dữ liệu asian_countries để lọc lại dữ liệu 51 quốc gia Châu Á

```python
asian = pd.read_csv('../data/raw_data/asian_countries.csv')
```

```python
asian
```

```python
asian_ast = pd.merge(ast, asian, on='ISO3')
```

```python
asian_ast = asian_ast.drop(['Unnamed: 0', 'Country_y'], axis=1)
```

```python
asian_ast = asian_ast.rename(columns={"Country_x":"Country"})
asian_ast
```
