```python
# Hàm tạo thêm các dòng mẫu với quốc gia tương ứng
def create_sample_rows(df, data):
    new_rows = []  # Danh sách chứa các dòng mẫu mới
    for sea, countries in data.items():
        for country in countries:
            # Sao chép các thông tin cột khác và thêm quốc gia mới cho mỗi đại dương
            new_row = df[df['Measure'] == sea].copy()  # Lọc các dòng có cùng 'sea'
            new_row['ISO3'] = country  # Thêm cột 'country' với giá trị quốc gia
            new_rows.append(new_row)

    # Nối các dòng mới vào DataFrame gốc
    new_df = pd.concat(new_rows, ignore_index=True)
    return new_df
```

```python
# Cập nhật DataFrame với quốc gia mới
df_updated = create_sample_rows(msl, data)
```

```python
asian_msl = pd.merge(df_updated, asian, on='ISO3')
```

```python
# Đổi vị trí cột `Country_y` lên trên `Country_x` và di chuyển `ObjectId` lên đầu
columns_order = [
    "ObjectId", "Country_y", "Country_x", "ISO2", "ISO3", "Indicator", 
    "Unit", "Source", "CTS Code", "CTS Name", "CTS Full Descriptor",
    "Measure", "Date", "Value", "Unnamed: 0.1", "Unnamed: 0"
]
asian_msl = asian_msl[columns_order]
```

```python
# Sắp xếp DataFrame theo cột ObjectId từ thấp đến cao
asian_msl = asian_msl.sort_values(by="ObjectId", ascending=True)

# Đặt lại chỉ số hàng sau khi sắp xếp
asian_msl = asian_msl.reset_index(drop=True)
```

```python
asian_msl = asian_msl.drop(['Unnamed: 0', 'Unnamed: 0.1', 'Country_x', 'ISO2', 'Source', 'CTS Code', 'CTS Name', 'CTS Full Descriptor'], axis=1)
```

```python
asian_msl = asian_msl.rename(columns={"Country_y":"Country"})
asian_msl
```

```python
# Chuyển đổi cột Date sang định dạng datetime và tạo cột Year
asian_msl['Date'] = pd.to_datetime(asian_msl['Date'], format='D%m/%d/%Y', errors='coerce')  # Định dạng ngày
asian_msl['Year'] = asian_msl['Date'].dt.year

# Nhóm theo Country và Year, tính giá trị trung bình của Value
result = asian_msl.groupby(['Country', 'Year', 'Indicator'])['Value'].mean().reset_index()

# Hiển thị kết quả
result
```

```python
asian_msl_final = pd.merge(result, asian_msl, on='Country', how='inner')
asian_msl_final
```

```python
asian_msl_final = asian_msl_final.drop(['Indicator_y', 'Value_y', 'Year_y',	'Date'], axis=1)
asian_msl_final
```

```python
asian_msl_final = asian_msl_final.drop('ObjectId', axis=1)
asian_msl_final = asian_msl_final.rename(columns={"Indicator_x":"Indicator"})
asian_msl_final = asian_msl_final.rename(columns={"Year_x":"Year"})
asian_msl_final = asian_msl_final.rename(columns={"Value_x":"Value"})
asian_msl_final
```

```python
# Đổi vị trí cột `Country_y` lên trên `Country_x` và di chuyển `ObjectId` lên đầu
columns_order = ["Country", "ISO3", "Indicator", "Unit", "Year", "Value", "Measure"]
asian_msl_final = asian_msl_final[columns_order]
```

```python
# Pivot dữ liệu
asian_msl_final = asian_msl_final.pivot_table(
    index=["Country", "ISO3", "Indicator", "Unit"],
    columns="Year",
    values="Value"
).reset_index()

# Hiển thị DataFrame với các năm là cột
asian_msl_final
```

```python
def preprocess_data(df, asian, output_path):

    # Step 1: Filter relevant columns
    df = df[['Country', 'ISO3', 'Indicator', 'Unit', 'F2018', 'F2019', 'F2020', 'F2021', 'F2022']]

    # Step 2: Merge with the 'asian' DataFrame on 'ISO3'
    asian_df = pd.merge(df, asian, on='ISO3')

    # Step 3: Drop unnecessary columns
    asian_df = asian_df.drop(['Unnamed: 0', 'Country_y'], axis=1, errors='ignore')

    # Step 4: Rename columns for consistency
    asian_df = asian_df.rename(columns={"Country_x": "Country"})

    # Step 5: Save the preprocessed data to a CSV file
    asian_df.to_csv(output_path, index=False)

    # Return the preprocessed DataFrame
    return asian_df
```
