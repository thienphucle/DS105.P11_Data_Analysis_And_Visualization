```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
data = pd.read_csv('../data/preprecessed_data/final_data.csv')
world_data =  data[data['ISO3']=='WLD']
world_data
```

```python
indicators = world_data['Indicator'].unique()
years = ['2018', '2019', '2020', '2021', '2022']
for indicator in indicators:
    # Filter data for the specific indicator
    indicator_data = world_data[world_data['Indicator'] == indicator]
    
    # Extract values for the y-axis
    values = indicator_data[years].values.flatten()
    
    # Plot line graph
    plt.figure(figsize=(10, 6))
    plt.plot(years, values, marker='o', linestyle='-')
    plt.title(f'Trend of {indicator}', fontsize=14)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel(indicator_data['Unit'].iloc[0], fontsize=12)
    plt.grid(True)
    plt.xticks(years)
    plt.tight_layout()
    plt.show()
```
